```python
# GENRATE RANDOM INTEGERS WITH MEAN AND SIGMA
mean=20
sigma=5
# generate the data and sort it
scatter_data_1=np.sort(np.random.normal(mean,sigma,500).astype(int))
scatter_data_2=np.sort(np.random.normal(mean,sigma,500).astype(int))

#scatterplot
plt.scatter(scatter_data_1,scatter_data_2)
```
